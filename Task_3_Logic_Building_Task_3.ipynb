```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
highest_name = []
highest_salary = 0

for i,j in employees.items():
    if j>highest_salary:
        highest_salary = j
        highest_name = [i]
    elif j == highest_salary:
        highest_name.append(i)
print(f"Top Performers Eligible for Bonus: {', '.join(highest_name)}(Score: {highest_salary})")
```

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
c=0
d=0

for i in emails:
    if '@gmail.com' in i:
        c+=1
    elif '@yahoo.com' in i:
        d+=1
print(f"gmail.com: {int((c/len(emails))*100)}%")
print(f"yahoo.com: {int((d/len(emails))*100)}%")      
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
total = sum(sales)
avg = total/len(sales)
margin_percentage = avg*1.30  # 2210
res = [(i+1,j)for i,j in enumerate(sales) if j > margin_percentage]
for i,j in res:
    print(f"Day {i}: {j}")
```
