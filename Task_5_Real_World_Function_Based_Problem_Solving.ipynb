```python
def analyze(likes):
  tot_likes = 0
  for like in likes:
    tot_likes += like
  if tot_likes >= 1000:
    status = "Viral Post"
  else:
    status = "Normal Engagement"
  print("Total Likes:", tot_likes)
  print("Post Status:", status)
#Example
likes = [200, 300, 150, 400]
analyze(likes)
```

```python
def check_rainfall(rainfall_data, req_level):
  total = 0
  for rain in rainfall_data:
    total += rain
  avg = total / len(rainfall_data)

  if avg >= req_level:
    status = "Adequate Rainfall"
  else:
    status = "Inadequate Rainfall"

  print("Average Rainfall:", avg)
  print("Rainfall Status:", status)

#Example
rainfall = [70, 75, 80, 63]
check_rainfall(rainfall, 70)
```

```python
def detect_duplicate(usernames):
  unique_users = set(usernames)

  if len(unique_users) != len(usernames):
    print("Duplicate Accounts Found: Yes")
  else:
    print("Duplicate Accounts Found: No")

#Example
users = ["john", "alice", "john", "mike"]
detect_duplicate(users)
```

```python
def filter_premium_crops(prices):
  premium= []
  for price in prices:
    if price > 2000:
      premium.append(price)
  print("Premium Crops:", premium)

#Example
crop_prices = [1500, 2500, 3200, 1800]
filter_premium_crops(crop_prices)
```

```python
def check_system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)


# Example
check_system_health(7)
```

```python
def check_attendance(attendance_list):
  total_classes = len(attendance_list)
  attended = 0

  for status in attendance_list:
    if status == 1:
      attended +=1
  percentage = (attended / total_classes) * 100

  if percentage >= 75:
    eligibility = "Eligible"
  else:
    eligibility = "Not Eligible"

  print ("Attendance Percentage:", percentage)
  print("Exam Eligibility:", eligibility)

#Example (1=present 0=absent)
attendance = [1,1,1,0,1]
check_attendance(attendance)
```

```python
#Smart electricty bill analyzer
def cal_electricity_bill(units):
  bill = 0

  if units <= 100:
    bill = units * 3
  elif units <= 200:
    bill = (100 * 3) + ((units - 100) * 5)
  else:
    bill = (100 * 3) + (100 * 5) +((units - 200) * 7)

  if bill < 500:
    usage = "Low Usage"
  elif bill <= 1500:
    usage = "Moderate Usage"
  else:
    usage = "High Usage"
  return bill, usage

#Example
total_bill, usage_status = cal_electricity_bill(250)
print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
